```python
import pandas as pd
import numpy as np
from fuzzywuzzy import fuzz, process

raw_path = "/KPMG_VI_New_raw_data_update_final.xlsx"
df_raw = pd.read_excel(raw_path)
df = df_raw.copy()
df.head()
```

## Duplicatas

```python
#Identificação
exact_duplicates = df.duplicated().sum()
```

```python
#Remoção
df_no_exect = df.drop_duplicates()
```

```python
#Detectar as aproximadas
subset_cols = ['CustormerID', 'CustomerName','Address']
sample = df_no_exect.head(100)
```

## Função para comparar strings

```python
def fuzzy_match (str1, str2, threshold=90):
  if pd.isna(str1) or pd.isna(str2):
    return False
  return fuzz.ratio(str1, str2) >= threshold
```

```python
#Encontrar duplicadas aproximadas
possible_dupes = []
for i in range(len(sample)):
    for j in range(i + 1, len(sample)):
        name1, name2 = sample.iloc[i]['Unnamed: 2'], sample.iloc[j]['Unnamed: 2'] # Changed 'CustomerName' to 'Unnamed: 2'
        if fuzzy_match(name1, name2):
            possible_dupes.append((name1, name2))

print(f"\nDuplicatas aproximadas detectadas: {len(possible_dupes)}")
for pair in possible_dupes[:5]:
    print("→", pair)
```

```python
if 'Unnamed: 3' in df_no_exect.columns:
    df_no_exect.sort_values(by='Unnamed: 3', ascending=False, inplace=True)
    df_final = df_no_exect.drop_duplicates(subset=['Unnamed: 2'], keep='first')
else:
    df_final = df_no_exect.copy()
```

```python
print(f"\n Linhas iniciais: {len(df)}")
print(f"Linhas após limpeza: {len(df_final)}")
print(f"Registros removidos: {len(df) - len(df_final)}")
```
