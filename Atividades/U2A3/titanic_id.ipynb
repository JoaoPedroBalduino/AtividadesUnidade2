```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings

raw_path = "/test.csv"
df_raw = pd.read_csv(raw_path)
df = df_raw.copy()
df.head()
```

## Identificar Problemas
-valores
ausentes, duplicatas e inconsistências.

```python
print(df.info())
print("\nNulos por coluna:\n", df.isna().sum())
print("\nDuplicados (linhas):", df.duplicated().sum())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mode()[0])
df['Cabin'] = df['Cabin'].fillna(df['Cabin'].mode()[0])
df['Fare'] = df['Fare'].fillna(df['Fare'].mode()[0])

print("\nValores ausentes após imputação:")
print(df.isna().sum())
```

```python
sex_mapping = {'male': 0, 'female': 1}
df['Sex'] = df['Sex'].map(sex_mapping)

embarked_mapping = {'S': 0, 'C': 1, 'Q': 2}
df['Embarked'] = df['Embarked'].map(embarked_mapping)

df['Age'] = df['Age'].astype(int)

print("\nTipos de dados após padronização:")
print(df.info())

print("\nDataFrame após padronização:")
display(df.head())
```
